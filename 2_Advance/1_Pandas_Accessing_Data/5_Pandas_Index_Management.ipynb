```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(15, random_state=42)
```

### Assign a name for the index column

```python
df.index.name = 'job_index'
```

# Three methods for index management

- df.reset_index()
- df.set_index()
- df.sort_index()

### df.reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
df_usa
```

```python
df_usa.reset_index(inplace=True)

df_usa
```

### df.set_index()

```python
df_usa.set_index('job_index', inplace=True)

df_usa
```

### df.sort_index()

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot
```

```python
median_pivot.sort_values(by=('median','salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
```

```python
median_pivot
```

```python
median_pivot.sort_values(by='job_title_short')
```
