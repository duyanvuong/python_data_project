### Question
In our project we want to see how in-demand skills are trending for data jobs. But before we do that let's look at:
- General trend of data jobs in demand. We're going to look at it by month
### Loading data

```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import calendar

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Demand of Data jobs in Vietnam

```python
df_vn = df.copy()
df_vn = df_vn[df_vn['job_country'] == 'Vietnam']
df_vn
```

### Create a month column

```python
df_vn['job_posted_month'] = df_vn['job_posted_date'].dt.strftime('%B')
df_vn
```

### Pivot table by JOB TITLE and MONTH

```python
df_vn_pivot = df_vn.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_vn_pivot
```

### Sorting values by month order

```python
df_vn_pivot.reset_index(inplace=True)
df_vn_pivot['month_no'] = pd.to_datetime(df_vn_pivot['job_posted_month'], format='%B').dt.month
df_vn_pivot.sort_values(by='month_no', inplace=True)
df_vn_pivot.set_index('job_posted_month', inplace=True)
df_vn_pivot.drop(columns='month_no', inplace=True)
df_vn_pivot
```

### Top 3 job title

```python
top_3 = df_vn['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
top_3
```

```python
df_vn_pivot[top_3]
```

```python
df_vn_pivot[top_3].plot(kind='line', xticks=range(12), figsize=(8,5))
plt.xticks(rotation=45, ha='center')
plt.xlabel('2023')
plt.ylabel('Job postings count')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.legend(title=False)
```
